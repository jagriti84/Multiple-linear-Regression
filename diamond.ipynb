```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataset
df = pd.read_csv("C:\Users\jagri\Downloads\archive (4).zip")

# Data Preprocessing
# Drop the 'Unnamed: 0' column
df.drop(columns=['Unnamed: 0'], inplace=True)

# Calculate 'Total Depth percentage' and drop 'carat', 'x', and 'y'
df['Total Depth percentage'] = round(2 * df['z'] / (df['x'] + df['y']) * 100, 2)
df.drop(columns=['carat', 'x', 'y'], inplace=True)

# Correlation Analysis
corr_matrix = df.corr()

# Data Visualization
plt.figure(figsize=(15, 5))
sns.histplot(x='depth', data=df)
plt.figure(figsize=(15, 5))
sns.histplot(x='table', data=df)
plt.figure(figsize=(15, 5))
sns.histplot(x='z', data=df)
plt.figure(figsize=(15, 5))
sns.histplot(x='price', data=df)

# Categorical Data Analysis
plt.figure(figsize=(15, 5))
sns.boxplot(x='cut', y='price', data=df)
plt.figure(figsize=(15, 5))
sns.boxplot(x='color', y='price', data=df)
plt.figure(figsize=(15, 5))
sns.boxplot(x='clarity', y='price', data=df)

# Regression Model Preparation
# Separate target variable and create dummy variables for categorical columns
y = df['price']
X = pd.get_dummies(df.drop(columns=['price']), drop_first=True)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2)

# Regression Model Building
regression = LinearRegression()
regression.fit(X_train, y_train)

# Model Evaluation
y_pred = regression.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(f"R-squared Score: {r2 * 100:.2f}%")
```
